# Fetch the top rated english movies in the US region using the TMDb API. From the result, print the first 10 movies which have original language as english. Also print their genres.

```python
# First we have to make a genre Dict
import requests

p=requests.get("https://api.themoviedb.org/3/genre/movie/list",params={"api_key":"19601e45727396ac951a4ac501939657"})

p.status_code
```

```python
import json

r=json.loads(p.text)
r
```

```python
Genre={}
for i in r["genres"]:
    Genre[i["id"]]=i["name"]
Genre
```

```python
# Now Finding Top rated movies

pq=requests.get("https://api.themoviedb.org/3/movie/top_rated",params={"api_key":"19601e45727396ac951a4ac501939657",})

pq.status_code
```

```python
rq=json.loads(pq.text)

rq
```

```python
# Now top 10 finding details for english movies

# Answer

k=0
for i in rq['results']:
    if i["original_language"]=="en":
        temp_id=i["id"]
        temp=requests.get("https://api.themoviedb.org/3/movie/"+str(temp_id),params={"api_key":"19601e45727396ac951a4ac501939657"})
        temp1=json.loads(temp.text)
        if len(temp1["production_countries"])==0:
            print(i["title"],end=" ")
            for j in temp1["genres"]:
                print(j["name"],end="")
            print()
            continue
        if temp1["production_countries"][0]["name"]=="United States of America" or temp1["production_countries"][0]["iso_3166_1"]=="GB":
            print(i["title"],end=" ")
            for j in temp1["genres"]:
                print(j["name"],end=" ")
            
            print()
```
