# Fetch the names of top 5 similar movies to 'Inception' from the TMDb API.

```python
import requests

r=requests.get("https://api.themoviedb.org/3/movie/top_rated",params={"api_key":"19601e45727396ac951a4ac501939657","original_language":"en","region":"US"})
r.status_code
```

```python
import json 
p=json.loads(r.text)
p
```

```python
for i in p["results"]:
    if i["original_language"]=="en":
        print(i["title"])
```
