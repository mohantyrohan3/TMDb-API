# Fetch the names of top 5 similar movies to 'Inception' from the TMDb API.

```python
# First we have to find the id of Inception

import requests

r=requests.get("https://api.themoviedb.org/3/search/movie",params={"api_key":"19601e45727396ac951a4ac501939657","query":"Inception"})
r.status_code
```

```python
import json

p=json.loads(r.text)
p
```

```python
# Inception id is 27205
rq=requests.get("https://api.themoviedb.org/3/movie/846190/similar",params={"api_key":"19601e45727396ac951a4ac501939657","movie_id":"846190"})
rq.status_code
```

```python
pq=json.loads(rq.text)
pq
```

```python
# Answer

for i in pq["results"]:
    print(i["title"],i["popularity"])
```
