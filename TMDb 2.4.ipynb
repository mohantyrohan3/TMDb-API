# Did James McAvoy play a role in the movie Deadpool 2. Print Yes or No.

```python
# First get id of Movie
import requests

r=requests.get("https://api.themoviedb.org/3/search/movie",params={"api_key":"19601e45727396ac951a4ac501939657","query":"Deadpool 2"})
r.status_code
```

```python
import json

p=json.loads(r.text)
p
```

```python
# Now Getting Details of the movie
ID=p["results"][0]["id"]

pq=requests.get("https://api.themoviedb.org/3/movie/383498/credits",params={"api_key":"19601e45727396ac951a4ac501939657"})
pq.status_code
```

```python
rq=json.loads(pq.text)
rq
```

```python
# Answer
Cast=[]
for i in rq["cast"]:
    Cast.append(i["name"])
if "James McAvoy" in Cast:
    print('Yes')
else:
    print("No")
```
